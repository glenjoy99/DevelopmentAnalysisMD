```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
data = pd.read_csv("./Maryland_Total_New_Housing_Units_Authorized_For_Construction__2000-2017.csv")
```

```python
data = data.set_index(data.Year)
```

```python
data = data.sort_index(0, ascending=True)
housingPG = data[data.columns[19]]
housingMO = data[data.columns[18]]
data.head()
```

```python
ax = plt.gca()
# plt.plot(data.Year, data[data.columns[3::]])
plt.plot(data.Year, housingPG, housingMO)
plt.title("Approved Housing Construction in PG and MoCo")
for label in ax.xaxis.get_ticklabels():
    label.set_rotation(30)
    
loc = plticker.MultipleLocator(base=1.0)
ax.xaxis.set_major_locator(loc)
fig = plt.gcf()
fig.set_size_inches(11,8)
plt.ylabel("Approved Housing Units")
plt.xlabel("Year")
plt.legend()
plt.show
```
